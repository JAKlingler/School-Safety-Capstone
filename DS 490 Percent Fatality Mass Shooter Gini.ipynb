```python
df = pd.read_csv(r'C:\Users\ryanb\Semester 8\DS 490 Capstone\DS 490 Report Writing\school_safety_all_data_merged.csv')
```

```python
df.head(1)
```

```python
df['Mass_Shooting'] = df['Number_Victims'].apply(lambda x: 1 if x >= 3 else 0)
```

```python
df['%_Fatality'] = df['Victims_Killed'] / df['Number_Victims']
```

```python
column_groups = {
    "ancestry": [col for col in df.columns if "Ancestry" in col],
    "immigration": [col for col in df.columns if "Immigration" in col],
    "race": [col for col in df.columns if "Race" in col],
    "school_enrollment": [col for col in df.columns if "School_Enrollment" in col],
    "disability_class": [col for col in df.columns if "Disability_Class" in col]
}

def gini(row, columns):
    values = []
    for value in row[columns]:
        try:
            values.append(float(value))
        except ValueError:
            # Skip non-numeric values
            pass
    
    if len(values) == 0:
        return np.nan  # Return NaN if no valid values are found
    
    sorted_values = sorted(values)
    n = len(values)
    cumulative_sum = np.cumsum(sorted_values)
    return (n + 1 - 2 * np.sum(cumulative_sum) / cumulative_sum[-1]) / n

for group, columns in column_groups.items():
    df[f'{group}_gini_score'] = df.apply(lambda row: gini(row, columns), axis=1)
```

```python
features = df.describe().columns.tolist()

column_to_exclude = '%_Fatality'

for group in column_groups:
    features = [col for col in features if col not in column_groups[group]]

if column_to_exclude in features:
    features.remove(column_to_exclude)

target = column_to_exclude

gini_score_columns = [f'{group}_gini_score' for group in column_groups]
features.extend(gini_score_columns)

df = df[features + [target]]
```

```python
columns_to_drop = 'Unnamed: 0'

df.drop(columns_to_drop, axis=1, inplace=True)
```
